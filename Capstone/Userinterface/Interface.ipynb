```python
def predict_points():
  print('This model requires Several inputs, please enter one of the options when prompted: ')

  print("\nOptions: 2022, 2023, 2024")
  year = int(input("Please input your chosen Season: "))
  if year < 2022 or year > 2024:
    return("Error! Invalid year please restart to try again")

  # Track Input
  track_info = track_input(year)

  # Driver Input
  driver_info = driver_input(year)

  # Start position input
  start = int(input("\nPlease input your chosen Driver's start position (1-20): "))
  if start < 1 or start > 20:
    return("Error! Start position must be between 1 and 20. Please restart to try again")

  # Checking that the functions run correctly
  # print(driver_info)
  # print(track_info)


  array_of_vals = [year,start]
  for x in track_info[1:23]:
    array_of_vals.append(x)

  for x in driver_info[1:21]:
    array_of_vals.append(x)

  # print(array_of_vals)

  new_df = pd.DataFrame(columns=c)
  new_df.loc[len(new_df)] = array_of_vals


  new_df[['prob_no_points','prob_points']] = model.predict_proba(new_df)
  print(  new_df[['prob_no_points','prob_points']])
  new_df['y_pred'] = np.where(new_df['prob_no_points']>.5, 0, 1)


  if new_df['y_pred'][0] == 1:
    print(f'\nWe predict that for your chosen values {driver_info[0]} will get points! :)')
  elif new_df['y_pred'][0] == 0:
    print(f'\nWe predict that for your chosen values {driver_info[0]} will not get points :(')
  else:
    return('There has been an Error please restart to try again.')


```
